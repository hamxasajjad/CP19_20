```python
## fix code
from tkinter import *
from tkinter.ttk import *
from PIL import ImageTk, Image
import csv
import time
import datetime
from pathlib import Path
import os

root = Tk()
root.geometry("841x601+251+73")
root.title("New Toplevel")
root.configure(background="#6c0e8b")
global combobox
combobox = StringVar()

def update_clock():
    images_names = os.listdir("DETECT")
    img=Image.open("1.png")
    img=img.resize((159, 176), Image.ANTIALIAS)
    img1 = ImageTk.PhotoImage(img)
    TLabel1.configure(image=img1)
    TLabel1.image=img1
    TLabel2.configure(text="STATUS: Finding")
    TLabel3.configure(text="Last Visit: ")
    TLabel4.configure(text="ID: ")
    TLabel5.configure(text="Total: ")
    tnam.set("")
    tcon.set("")
    temail.set("")


    
    for image_name in images_names:
        my_file = Path("DETECT/"+ image_name)
        img=Image.open(my_file)
        img=img.resize((159, 176), Image.ANTIALIAS)
        img1 = ImageTk.PhotoImage(img)
        TLabel1.configure(image=img1)
        TLabel1.image=img1
        TLabel2.configure(text="STATUS: Found " + image_name[:(len(image_name)-4)])
        if image_name[:(len(image_name)-4)] == "UNKNOWN":
            TLabel2.configure(text="STATUS: UNKNOWN")
        else:
            tnam.set(image_name[:(len(image_name)-4)])
            listdata=[]    
            with open('data.csv', mode='r') as csv_file:
                csv_reader = csv.reader(csv_file)
                for row in csv_reader:
                    if row !=  [] :
                        listdata.append(row) 
            for i in range(len(listdata)):
                if listdata[i][0]==tnam.get():
                    tcon.set(listdata[i][1])
                    temail.set(listdata[i][2])
                    TLabel3.configure(text="Last Visit: "+listdata[i][6])
                    TLabel4.configure(text="ID: "+listdata[i][3])
                    TLabel5.configure(text="Total: "+listdata[i][5])


                    
    root.after(10000, update_clock)
    
######################## classes

class AutoScroll(object):
    def __init__(self, master):
        try:
            vsb =  Scrollbar(master, orient='vertical', command=self.yview)
        except:
            pass
        hsb =  Scrollbar(master, orient='horizontal', command=self.xview)

        #self.configure(yscrollcommand=_autoscroll(vsb),
        #    xscrollcommand=_autoscroll(hsb))
        try:
            self.configure(yscrollcommand=self._autoscroll(vsb))
        except:
            pass
        self.configure(xscrollcommand=self._autoscroll(hsb))

        self.grid(column=0, row=0, sticky='nsew')
        try:
            vsb.grid(column=1, row=0, sticky='ns')
        except:
            pass
        hsb.grid(column=0, row=1, sticky='ew')

        master.grid_columnconfigure(0, weight=1)
        master.grid_rowconfigure(0, weight=1)

        # Copy geometry methods of master  (taken from ScrolledText.py)
        methods = list(Pack.__dict__) + list(Grid.__dict__) + list(Place.__dict__)

        for meth in methods:
            if meth[0] != '_' and meth not in ('config', 'configure'):
                setattr(self, meth, getattr(master, meth))

    @staticmethod
    def _autoscroll(sbar):
        '''Hide and show scrollbar as needed.'''
        def wrapped(first, last):
            first, last = float(first), float(last)
            if first <= 0 and last >= 1:
                sbar.grid_remove()
            else:
                sbar.grid()
            sbar.set(first, last)
        return wrapped

    def __str__(self):
        return str(self.master)

def _create_container(func):
    '''Creates a ttk Frame with a given master, and use this new frame to
    place the scrollbars and the widget.'''
    def wrapped(cls, master, **kw):
        container =  Frame(master)
        return func(cls, container, **kw)
    return wrapped
    
    
class ScrolledTreeView(AutoScroll, Treeview):
    @_create_container
    def __init__(self, master, **kw):
        Treeview.__init__(self, master, **kw)
        AutoScroll.__init__(self, master)


############### functions

def b1f():
    
    Scrolledtreeview1.insert('', 'end', text="" , values=(sid.get(),combobox.get(),quantity.get(),price.get()))
       
def b2f():
   
    global myData
    myData.append([tnam.get(),tcon.get(),temail.get(),sid.get(),combobox.get(),quantity.get(),date1.get(),time1.get(),price.get()])
    myFile = open('data.csv', 'a')
    with myFile:
        writer = csv.writer(myFile)
        writer.writerows(myData)

    tnam.set("")
    tcon.set("")
    temail.set("")
    sid.set("")
    combobox.set("")
    quantity.set("") 
    date1.set("") 
    time1.set("") 
    price.set("") 
    for i in Scrolledtreeview1.get_children():
        Scrolledtreeview1.delete(i)
    images_names = os.listdir("DETECT")
    for image_name in images_names:
        my_file = Path("DETECT/"+ image_name)
        if os.path.exists(my_file):
            os.remove(my_file)
    
   
    
myData=[]
    
date1=StringVar()
time1=StringVar()
    
date1.set(time.strftime('%d/%m/%y'))
time1.set(time.strftime('%I:%M:%S'))
    
    
    
    

######################### code to paste

_bgcolor = '#d9d9d9'  # X11 color: 'gray85'
_fgcolor = '#000000'  # X11 color: 'black'
_compcolor = '#d9d9d9' # X11 color: 'gray85'
_ana1color = '#d9d9d9' # X11 color: 'gray85' 
_ana2color = '#d9d9d9' # X11 color: 'gray85' 
font12 = "-family {Tempus Sans ITC} -size 9 -weight bold "  \
            "-slant roman -underline 0 -overstrike 0"
font13 = "-family {Tempus Sans ITC} -size 11 -weight bold "  \
            "-slant roman -underline 0 -overstrike 0"
font14 = "-family {Tempus Sans ITC} -size 10 -weight bold "  \
            "-slant roman -underline 0 -overstrike 0"



style = Style()

sid = StringVar()
tnam = StringVar()
tcon = StringVar()
temail = StringVar()
date = StringVar()
article = StringVar()
quantity = StringVar()
price = StringVar()


   
if sys.platform == "win32":
            style.theme_use('winnative')
style.configure('.',background=_bgcolor)
style.configure('.',foreground=_fgcolor)
style.configure('.',font="TkDefaultFont")
style.map('.',background=[('selected', _compcolor), ('active',_ana2color)])

Frame1 = Frame(root)
Frame1.place(relx=0.02, rely=0.03, relheight=0.92, relwidth=0.64)
Frame1.configure(relief=GROOVE)
Frame1.configure(borderwidth="2")
Frame1.configure(relief=GROOVE)
#Frame1.configure(background="#efefef")
Frame1.configure(width=535)

text1 = Entry(Frame1)
text1.place(relx=0.21, rely=0.09, relheight=0.04, relwidth=0.24)
text1.configure(takefocus="",textvariable=sid)
text1.configure(cursor="ibeam")


text4 = Entry(Frame1)
text4.place(relx=0.69, rely=0.16, relheight=0.04, relwidth=0.24)
text4.configure(takefocus="",textvariable=time1)
text4.configure(cursor="ibeam")

text3 = Entry(Frame1)
text3.place(relx=0.69, rely=0.09, relheight=0.04, relwidth=0.24)
text3.configure(takefocus="",textvariable=date1)
text3.configure(cursor="ibeam")

TCombo1 = Entry(Frame1)
TCombo1.place(relx=0.21, rely=0.16, relheight=0.04, relwidth=0.23)
TCombo1.configure(takefocus="",textvariable=combobox)
TCombo1.configure(cursor="ibeam")

text2 = Entry(Frame1)
text2.place(relx=0.21, rely=0.25, relheight=0.04, relwidth=0.24)
text2.configure(takefocus="",textvariable= quantity)
text2.configure(cursor="ibeam")

text5 = Entry(Frame1)
text5.place(relx=0.69, rely=0.25, relheight=0.04, relwidth=0.24)
text5.configure(takefocus="",textvariable= price)
text5.configure(cursor="ibeam")

b1 = Button(Frame1)
b1.place(relx=0.21, rely=0.34, height=25, width=76)
b1.configure(takefocus="")
b1.configure(text='''ADD''')
b1.configure(command=b1f)


style.configure('Treeview.Heading',  font="TkDefaultFont")

Scrolledtreeview1 = ScrolledTreeView(Frame1)
Scrolledtreeview1.place(relx=0.02, rely=0.49, relheight=0.32
   , relwidth=0.96)
Scrolledtreeview1.configure(columns=" Col1 Col2 Col3 Col4")
Scrolledtreeview1.column("#0",width="78")
Scrolledtreeview1.column("#0",minwidth="20")
Scrolledtreeview1.column("#0",stretch="1")
Scrolledtreeview1.column("#0",anchor="w")
Scrolledtreeview1.heading("Col1",text="Sid")
Scrolledtreeview1.heading("Col1",anchor="center")
Scrolledtreeview1.column("Col1",width="103")
Scrolledtreeview1.column("Col1",minwidth="20")
Scrolledtreeview1.column("Col1",stretch="1")
Scrolledtreeview1.column("Col1",anchor="w")
Scrolledtreeview1.heading("Col2",text="Article")
Scrolledtreeview1.heading("Col2",anchor="center")
Scrolledtreeview1.column("Col2",width="103")
Scrolledtreeview1.column("Col2",minwidth="20")
Scrolledtreeview1.column("Col2",stretch="1")
Scrolledtreeview1.column("Col2",anchor="w")
Scrolledtreeview1.heading("Col3",text="Quantity")
Scrolledtreeview1.heading("Col3",anchor="center")
Scrolledtreeview1.column("Col3",width="104")
Scrolledtreeview1.column("Col3",minwidth="20")
Scrolledtreeview1.column("Col3",stretch="1")
Scrolledtreeview1.column("Col3",anchor="w")
Scrolledtreeview1.heading("Col4",text="Price")
Scrolledtreeview1.heading("Col4",anchor="center")
Scrolledtreeview1.column("Col4",width="104")
Scrolledtreeview1.column("Col4",minwidth="20")
Scrolledtreeview1.column("Col4",stretch="1")
Scrolledtreeview1.column("Col4",anchor="w")

Label4 = Label(Frame1)
Label4.place(relx=0.04, rely=0.09, height=31, width=64)
Label4.configure(background="#d9d9d9")
#Label4.configure(disabledforeground="#a3a3a3")
Label4.configure(font=font13)
Label4.configure(foreground="#000000")
Label4.configure(text='''SLIP ID:''')
Label4.configure(width=64)

Label5 = Label(Frame1)
Label5.place(relx=0.04, rely=0.16, height=25, width=74)
Label5.configure(background="#d9d9d9")
#Label5.configure(disabledforeground="#a3a3a3")
Label5.configure(font=font13)
Label5.configure(foreground="#000000")
Label5.configure(text='''ARTICLE:''')

Label6 = Label(Frame1)
Label6.place(relx=0.04, rely=0.23, height=41, width=74)
Label6.configure(background="#d9d9d9")
#Label6.configure(disabledforeground="#a3a3a3")
Label6.configure(font=font14)
Label6.configure(foreground="#000000")
Label6.configure(text='''QUANTITY:''')
Label6.configure(width=74)

Label7 = Label(Frame1)
Label7.place(relx=0.52, rely=0.09, height=21, width=44)
Label7.configure(background="#d9d9d9")
#Label7.configure(disabledforeground="#a3a3a3")
Label7.configure(font=font13)
Label7.configure(foreground="#000000")
Label7.configure(text='''DATE:''')
Label7.configure(width=44)

Label8 = Label(Frame1)
Label8.place(relx=0.52, rely=0.16, height=25, width=48)
Label8.configure(background="#d9d9d9")
#Label8.configure(disabledforeground="#a3a3a3")
Label8.configure(font=font13)
Label8.configure(foreground="#000000")
Label8.configure(text='''TIME:''')

Label9 = Label(Frame1)
Label9.place(relx=0.52, rely=0.25, height=25, width=54)
Label9.configure(background="#d9d9d9")
#Label9.configure(disabledforeground="#a3a3a3")
Label9.configure(font=font13)
Label9.configure(foreground="#000000")
Label9.configure(text='''PRICE:''')

b2 = Button(Frame1)
b2.place(relx=0.22, rely=0.85, height=25, width=76)
b2.configure(takefocus="")
b2.configure(text='''SAVE''')
b2.configure(command=b2f)

TFrame1 = Frame(root)
TFrame1.place(relx=0.68, rely=0.03, relheight=0.92, relwidth=0.28)
TFrame1.configure(relief=GROOVE)
TFrame1.configure(borderwidth="2")
TFrame1.configure(relief=GROOVE)
TFrame1.configure(width=235)

TEntry6 = Entry(TFrame1)
TEntry6.place(relx=0.34, rely=0.74, relheight=0.04, relwidth=0.54)
TEntry6.configure(takefocus="",textvariable=tnam)
TEntry6.configure(cursor="ibeam")

TEntry7 = Entry(TFrame1)
TEntry7.place(relx=0.34, rely=0.83, relheight=0.04, relwidth=0.54)
TEntry7.configure(takefocus="",textvariable=tcon)
TEntry7.configure(cursor="ibeam")

TEntry8 = Entry(TFrame1)
TEntry8.place(relx=0.34, rely=0.9, relheight=0.04, relwidth=0.54)
TEntry8.configure(takefocus="",textvariable=temail)
TEntry8.configure(cursor="ibeam")

TLabelframe1 = Labelframe(TFrame1)
TLabelframe1.place(relx=0.09, rely=0.02, relheight=0.55
   , relwidth=0.77)
TLabelframe1.configure(width=180)

##################################
#img=Image.open("r.png")
#img=img.resize((159, 176), Image.ANTIALIAS)
#img1 = ImageTk.PhotoImage(img)
TLabel1 = Label(TLabelframe1)
TLabel1.place(relx=0.0, rely=0.07, height=159, width=176, y=-12)
TLabel1.configure(text='''customer image''')


TLabel2 = Label(TLabelframe1)
TLabel2.place(relx=0.0, rely=0.62, height=29, width=336, y=-12)
TLabel2.configure(background="#efefef")
TLabel2.configure(foreground="#000000")
TLabel2.configure(font=font14)
TLabel2.configure(relief=FLAT)
TLabel2.configure(text='''STATUS:''')
TLabel2.configure(width=336)

TLabel3 = Label(TLabelframe1)
TLabel3.place(relx=0.0, rely=0.75, height=29, width=186, y=-12)
TLabel3.configure(background="#efefef")
TLabel3.configure(foreground="#000000")
TLabel3.configure(font=font14)
TLabel3.configure(relief=FLAT)
TLabel3.configure(text='''LAST VISIT :''')
TLabel3.configure(width=186)

TLabel4 = Label(TLabelframe1)
TLabel4.place(relx=0.0, rely=0.89, height=29, width=186, y=-12)
TLabel4.configure(background="#efefef")
TLabel4.configure(foreground="#000000")
TLabel4.configure(font=font14)
TLabel4.configure(relief=FLAT)
TLabel4.configure(text='''CUSTOMER ID:''')
TLabel4.configure(width=186)

Label1 = Label(TFrame1)
Label1.place(relx=0.09, rely=0.74, height=21, width=44)
Label1.configure(background="#d9d9d9")
#Label1.configure(disabledforeground="#a3a3a3")
Label1.configure(font=font14)
Label1.configure(foreground="#000000")
Label1.configure(text='''NAME:''')
Label1.configure(width=44)

Label2 = Label(TFrame1)
Label2.place(relx=0.04, rely=0.83, height=21, width=64)
Label2.configure(background="#d9d9d9")
#Label2.configure(disabledforeground="#a3a3a3")
Label2.configure(font=font12)
Label2.configure(foreground="#000000")
Label2.configure(text='''CONTACT:''')
Label2.configure(width=64)

Label3 = Label(TFrame1)
Label3.place(relx=0.04, rely=0.9, height=31, width=54)
Label3.configure(background="#d9d9d9")
#Label3.configure(disabledforeground="#a3a3a3")
Label3.configure(font=font12)
Label3.configure(foreground="#000000")
Label3.configure(text='''EMAIL:''')
Label3.configure(width=54)

TLabel5 = Label(TFrame1)
TLabel5.place(relx=0.09, rely=0.58, height=39, width=186)
TLabel5.configure(background="#efefef")
TLabel5.configure(foreground="#000000")
TLabel5.configure(font=font14)
TLabel5.configure(relief=FLAT)
TLabel5.configure(text='''TOTAL PURCHASE:''')
TLabel5.configure(width=186)

menubar = Menu(root,font="TkMenuFont",bg=_bgcolor,fg=_fgcolor)
root.configure(menu = menubar)
#update_clock()

################ end paste code

root.mainloop()

```
